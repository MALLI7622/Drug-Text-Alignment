```python
# ==== 3) Helper: mean-pool last hidden state with attention mask ====
def mean_pool(last_hidden_state, attention_mask):
    # last_hidden_state: [B, T, H], attention_mask: [B, T]
    mask = attention_mask.unsqueeze(-1).type_as(last_hidden_state)
    summed = (last_hidden_state * mask).sum(dim=1)
    counts = mask.sum(dim=1).clamp(min=1e-9)
    return summed / counts
```

```python
@torch.no_grad()
def encode_texts(texts, tok, model, max_length=128, batch_size=32):
    reps = []
    for i in tqdm(range(0, len(texts), batch_size), desc="Encoding texts"):
        batch = texts[i:i+batch_size]
        enc = tok(batch, padding=True, truncation=True, max_length=max_length, return_tensors="pt").to(device)
        out = model(**enc)
        emb = mean_pool(out.last_hidden_state, enc["attention_mask"])
        reps.append(emb.detach().cpu())
    return torch.cat(reps, dim=0)
```

```python
@torch.no_grad()
def encode_smiles(smiles, tok, model, max_length=128, batch_size=32):
    reps = []
    for i in tqdm(range(0, len(smiles), batch_size), desc="Encoding SMILES"):
        batch = smiles[i:i+batch_size]
        enc = tok(batch, padding=True, truncation=True, max_length=128, return_tensors="pt").to(device)
        out = model(**enc)
        emb = mean_pool(out.last_hidden_state, enc["attention_mask"])
        reps.append(emb.detach().cpu())
    return torch.cat(reps, dim=0)
```

```python
# ==== 4) Encode ====
E_text   = encode_texts(text_list, text_tok, text_model, max_length=64, batch_size=64)     # [N, H_t]
E_smiles = encode_smiles(smiles_list, chem_tok, chem_model, max_length=128, batch_size=64) # [N, H_s]
```

```python
import torch, torch.nn as nn, torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import math, random

torch.manual_seed(0); np.random.seed(0); random.seed(0)

# Cache the frozen CLS embeddings from step 1 (or 2)
Zt = E_text.clone()      # [N, Ht]
Zs = E_smiles.clone()    # [N, Hs]

class PairDS(Dataset):
    def __init__(self, Zt, Zs):
        self.Zt = Zt; self.Zs = Zs
    def __len__(self): return len(self.Zt)
    def __getitem__(self, i): return self.Zt[i], self.Zs[i]

ds = PairDS(Zt, Zs)
dl = DataLoader(ds, batch_size=256, shuffle=True, drop_last=True)

d_text = Zt.shape[1]; d_smiles = Zs.shape[1]; d_sh = 256

proj_text   = nn.Linear(d_text, d_sh).to(device)
proj_smiles = nn.Linear(d_smiles, d_sh).to(device)

opt = optim.AdamW(list(proj_text.parameters())+list(proj_smiles.parameters()), lr=1e-3, weight_decay=1e-4)
T = 0.07  # temperature

def contrastive_step(bt, bs):
    # bt, bs: [B, d]
    bt = F.normalize(proj_text(bt.to(device)), dim=1)
    bs = F.normalize(proj_smiles(bs.to(device)), dim=1)
    logits = (bt @ bs.T) / T     # [B, B]
    labels = torch.arange(bt.size(0), device=bt.device)
    loss_i = F.cross_entropy(logits, labels)
    loss_j = F.cross_entropy(logits.T, labels)
    loss = (loss_i + loss_j) / 2
    return loss

epochs = 20  # ~5–10 minutes on Colab; reduce to 3 if needed
for ep in range(1, epochs+1):
    proj_text.train(); proj_smiles.train()
    running = 0.0; steps = 0
    for bt, bs in dl:
        opt.zero_grad()
        loss = contrastive_step(bt, bs)
        loss.backward()
        opt.step()
        running += loss.item(); steps += 1
    print(f"Epoch {ep}/{epochs}  loss={running/steps:.4f}")

# Evaluate with learned heads
@torch.no_grad()
def eval_with_heads(Zt, Zs):
    Bt = F.normalize(proj_text(Zt.to(device)), dim=1).cpu()
    Bs = F.normalize(proj_smiles(Zs.to(device)), dim=1).cpu()
    S = Bt @ Bs.T
    gt = np.arange(S.shape[0])
    top1 = S.argmax(dim=1).numpy()
    recall1 = (top1 == gt).mean()
    ranks = []
    for i in range(S.shape[0]):
        order = torch.argsort(S[i], descending=True).numpy()
        rank_pos = int(np.where(order == i)[0][0]) + 1
        ranks.append(1.0 / rank_pos)
    mrr = float(np.mean(ranks))
    return recall1, mrr

rec1, mrr = eval_with_heads(Zt, Zs)
print(f"[After tiny contrastive bridge] Recall@1: {rec1:.3f} | MRR: {mrr:.3f}")
```

```python
import numpy as np, torch, torch.nn.functional as F
from rdkit import Chem
from rdkit.Chem import AllChem
from torch.utils.data import DataLoader, Dataset
import torch.nn as nn, torch.optim as optim
import pandas as pd, random
import warnings; warnings.filterwarnings("ignore")

device = "cuda" if torch.cuda.is_available() else "cpu"
df = pd.read_csv("chembl_mechanisms_enriched.csv").dropna(subset=["smiles","text_rich"])
df = df[df["text_rich"].str.len() > 20].reset_index(drop=True)

# 1) Compute ECFP4 (2048 bits)
def ecfp4_bitvect(smi, nbits=2048, radius=2):
    m = Chem.MolFromSmiles(smi)
    if m is None: return np.zeros(nbits, dtype=np.float32)
    bv = AllChem.GetMorganFingerprintAsBitVect(m, radius, nBits=nbits)
    arr = np.zeros((nbits,), dtype=np.int8)
    Chem.DataStructs.ConvertToNumpyArray(bv, arr)
    return arr.astype(np.float32)

X_mol = np.stack([ecfp4_bitvect(s) for s in df["smiles"].tolist()], axis=0)  # [N,2048]

# 2) Get text embeddings (reuse your best text encoder from earlier steps)
#    Assume you already have a function cls_encode(...) and loaded text_tok/text_model
from transformers import AutoTokenizer, AutoModel
text_model_name = "pritamdeka/S-Biomed-Roberta-snli-multinli-stsb"  # or your best
text_tok = AutoTokenizer.from_pretrained(text_model_name)
text_model = AutoModel.from_pretrained(text_model_name).to(device).eval()

@torch.no_grad()
def cls_encode(texts, tok, model, max_length=96, bs=64):
    outs = []
    for i in range(0, len(texts), bs):
        batch = texts[i:i+bs]
        enc = tok(batch, padding=True, truncation=True, max_length=max_length, return_tensors="pt").to(device)
        out = model(**enc).last_hidden_state[:,0,:]
        outs.append(out.detach().cpu())
    return torch.cat(outs, dim=0)

Z_text = cls_encode(df["text_rich"].tolist(), text_tok, text_model)  # [N, Ht]

# 3) Contrastive training: frozen text, learn only mol projector
class PairDS(Dataset):
    def __init__(self, Zt, Xm):
        self.Zt = torch.tensor(Zt, dtype=torch.float32)
        self.Xm = torch.tensor(Xm, dtype=torch.float32)
    def __len__(self): return len(self.Zt)
    def __getitem__(self, i): return self.Zt[i], self.Xm[i]

ds = PairDS(Z_text.numpy(), X_mol)
dl = DataLoader(ds, batch_size=512, shuffle=True, drop_last=True)

d_text = Z_text.shape[1]; d_fp = X_mol.shape[1]; d_sh = 256
proj_text = nn.Linear(d_text, d_sh).to(device)     # learn small head (or fix to identity via nn.Identity())
proj_mol  = nn.Linear(d_fp,   d_sh).to(device)

opt = optim.AdamW(list(proj_text.parameters())+list(proj_mol.parameters()), lr=1e-3, weight_decay=1e-4)
T = 0.07

def step(bt, bm):
    bt = F.normalize(proj_text(bt.to(device)), dim=1)
    bm = F.normalize(proj_mol (bm.to(device)), dim=1)
    logits = bt @ bm.T / T
    labels = torch.arange(bt.size(0), device=bt.device)
    loss = (F.cross_entropy(logits, labels) + F.cross_entropy(logits.T, labels)) / 2
    return loss

for ep in range(1, 21):  # start with 3 epochs (fast)
    tot = n = 0
    for bt, bm in dl:
        opt.zero_grad()
        loss = step(bt, bm)
        loss.backward()
        opt.step()
        tot += loss.item(); n += 1
    print(f"ECFP4 bridge epoch {ep} loss={tot/n:.4f}")

# 4) Evaluate Recall@1/MRR
@torch.no_grad()
def eval_recall_mrr(Zt, Xm):
    Bt = F.normalize(proj_text(torch.tensor(Zt, dtype=torch.float32).to(device)), dim=1).cpu()
    Bm = F.normalize(proj_mol (torch.tensor(Xm, dtype=torch.float32).to(device)), dim=1).cpu()
    S = Bt @ Bm.T
    gt = np.arange(S.shape[0])
    top1 = S.argmax(dim=1).numpy()
    recall1 = (top1 == gt).mean()
    ranks = []
    for i in range(S.shape[0]):
        order = torch.argsort(S[i], descending=True).numpy()
        r = int(np.where(order == i)[0][0]) + 1
        ranks.append(1.0/r)
    return recall1, float(np.mean(ranks))

rec1, mrr = eval_recall_mrr(Z_text.numpy(), X_mol)
print(f"[ECFP4 bridge] Recall@1: {rec1:.3f} | MRR: {mrr:.3f}")
```

```python
import numpy as np, pandas as pd, torch

# After you have a similarity matrix S between text and molecule representations:
# Here we show how to compute group-wise Recall@1 by target_chembl_id
def grouped_recall_at1(S, meta_df, group_col="target_chembl_id"):
    # S: [N,N] similarities; i-th text paired with i-th molecule (same ordering)
    assert S.shape[0] == S.shape[1] == len(meta_df)
    groups = meta_df[group_col].fillna("UNK").astype(str).values
    idx_by_group = {}
    for i, g in enumerate(groups):
        idx_by_group.setdefault(g, []).append(i)

    hits = total = 0
    for g, idxs in idx_by_group.items():
        if len(idxs) < 3:  # skip tiny groups
            continue
        sub = S[np.ix_(idxs, idxs)]  # restrict to group
        top1 = sub.argmax(axis=1)
        hits += (top1 == np.arange(len(idxs))).sum()
        total += len(idxs)
    return hits/total if total else 0.0

```

```python


import os, math, random, numpy as np, pandas as pd, torch, torch.nn as nn, torch.nn.functional as F
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Scaffolds import MurckoScaffold
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import Dataset, DataLoader

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# =============================
# Config
# =============================
CSV_PATH = "chembl_mechanisms_enriched.csv"   # must have: smiles, text_rich, target_chembl_id (plus any others)
TEXT_MODEL = "pritamdeka/S-Biomed-Roberta-snli-multinli-stsb"  # swap if needed
MAX_TXT_LEN = 96
BATCH_SIZE = 512
EPOCHS = 21                  # 6-10 works well on Colab
LR = 1e-3
WEIGHT_DECAY = 1e-4
TEMP = 0.07                 # InfoNCE temperature
SHARED_D = 256              # projection size
ECFP_BITS = 2048
ECFP_RADIUS = 2
SEED = 0

# Evaluation toggles
REMOVE_DRUG_NAME = False    # leakage control: if your text_rich contains "Drug: {pref_name}", set True to strip it
MIN_GROUP_SIZE = 3          # for grouped Recall@1
TEST_ONLY_TOPK = None       # e.g., 500 to speed eval; None = all

# =============================
# Utils
# =============================
def set_seed(seed=0):
    random.seed(seed); np.random.seed(seed); torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)

def ecfp4_bitvect(smi, nbits=2048, radius=2):
    m = Chem.MolFromSmiles(smi)
    if m is None:
        return np.zeros(nbits, dtype=np.float32)
    bv = AllChem.GetMorganFingerprintAsBitVect(m, radius, nBits=nbits)
    arr = np.zeros((nbits,), dtype=np.int8)
    Chem.DataStructs.ConvertToNumpyArray(bv, arr)
    return arr.astype(np.float32)

def murcko(sm):
    m = Chem.MolFromSmiles(sm)
    if m is None: return None
    core = MurckoScaffold.GetScaffoldForMol(m)
    return Chem.MolToSmiles(core) if core else None

@torch.no_grad()
def cls_encode(texts, tok, model, max_length=96, bs=64):
    outs = []
    for i in tqdm(range(0, len(texts), bs), desc="Encode text (CLS)"):
        batch = texts[i:i+bs]
        enc = tok(batch, padding=True, truncation=True, max_length=max_length, return_tensors="pt").to(device)
        out = model(**enc).last_hidden_state[:,0,:]  # CLS token
        outs.append(out.detach().cpu())
    return torch.cat(outs, dim=0)

def build_group_index(series):
    """Map group label -> list of indices"""
    groups = series.fillna("UNK").astype(str).values
    idx_by_group = {}
    for i, g in enumerate(groups):
        idx_by_group.setdefault(g, []).append(i)
    return idx_by_group

def grouped_recall_at1(S_np, meta_df, group_col="target_chembl_id", min_group=3):
    """Recall@1 computed within each group, then micro-averaged."""
    assert S_np.shape[0] == S_np.shape[1] == len(meta_df)
    idx_by_group = build_group_index(meta_df[group_col])
    hits = total = 0
    for g, idxs in idx_by_group.items():
        if len(idxs) < min_group:
            continue
        sub = S_np[np.ix_(idxs, idxs)]
        top1 = sub.argmax(axis=1)
        hits += (top1 == np.arange(len(idxs))).sum()
        total += len(idxs)
    return hits/total if total else 0.0

# =============================
# 1) Load & prepare data
# =============================
set_seed(SEED)
df = pd.read_csv(CSV_PATH)
need_cols = {"smiles","text_rich","target_chembl_id"}
missing = need_cols - set(df.columns)
if missing:
    raise ValueError(f"CSV missing required columns: {missing}")

df = df.dropna(subset=["smiles","text_rich"]).copy()

if REMOVE_DRUG_NAME:
    # crude removal of "Drug: X." pattern if present in text_rich
    df["text_rich"] = df["text_rich"].str.replace(r"Drug:\s*[^.]+\. ?", "", regex=True).str.strip()

# compute Murcko scaffolds and split
print("Computing Murcko scaffolds...")
df["scaffold"] = df["smiles"].apply(murcko)
df = df.dropna(subset=["scaffold"]).reset_index(drop=True)
scaffolds = df["scaffold"].unique()
np.random.seed(SEED); np.random.shuffle(scaffolds)
cut = int(0.8*len(scaffolds))
train_scafs = set(scaffolds[:cut])
test_scafs  = set(scaffolds[cut:])

train_df = df[df["scaffold"].isin(train_scafs)].reset_index(drop=True)
test_df  = df[df["scaffold"].isin(test_scafs)].reset_index(drop=True)
print(f"Split sizes -> train: {len(train_df)} | test: {len(test_df)}")

# Optional: limit test size for speed
if TEST_ONLY_TOPK is not None and len(test_df) > TEST_ONLY_TOPK:
    test_df = test_df.sample(TEST_ONLY_TOPK, random_state=SEED).reset_index(drop=True)
    print(f"Test subset for speed: {len(test_df)}")

# =============================
# 2) Build features: ECFP4 (mol) + Text CLS (frozen)
# =============================
print("Computing ECFP4 fingerprints...")
X_mol_train = np.stack([ecfp4_bitvect(s, ECFP_BITS, ECFP_RADIUS) for s in tqdm(train_df["smiles"].tolist())])
X_mol_test  = np.stack([ecfp4_bitvect(s, ECFP_BITS, ECFP_RADIUS) for s in tqdm(test_df["smiles"].tolist())])

print("Encoding text embeddings...")
text_tok = AutoTokenizer.from_pretrained(TEXT_MODEL, use_fast=True)
text_model = AutoModel.from_pretrained(TEXT_MODEL).to(device).eval()
Z_text_train = cls_encode(train_df["text_rich"].tolist(), text_tok, text_model, max_length=MAX_TXT_LEN, bs=64)
Z_text_test  = cls_encode(test_df["text_rich"].tolist(),  text_tok, text_model, max_length=MAX_TXT_LEN, bs=64)

# =============================
# 3) Dataset/Loader for training bridge
# =============================
class PairDataset(Dataset):
    def __init__(self, Z_text, X_mol, group_labels):
        self.Zt = torch.tensor(Z_text, dtype=torch.float32)
        self.Xm = torch.tensor(X_mol,  dtype=torch.float32)
        self.grp = np.array(group_labels).astype(str)
    def __len__(self): return len(self.Zt)
    def __getitem__(self, i):
        return self.Zt[i], self.Xm[i], self.grp[i], i

train_ds = PairDataset(Z_text_train.numpy(), X_mol_train, train_df["target_chembl_id"].tolist())
train_dl = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)

# =============================
# 4) Thin contrastive bridge with hard-neg weighting + margin loss
# =============================
d_text = Z_text_train.shape[1]
d_fp   = X_mol_train.shape[1]

proj_text = nn.Linear(d_text, SHARED_D).to(device)
proj_mol  = nn.Linear(d_fp,   SHARED_D).to(device)

opt = torch.optim.AdamW(list(proj_text.parameters())+list(proj_mol.parameters()), lr=LR, weight_decay=WEIGHT_DECAY)

def build_same_target_weights(grps):
    """Return weight matrix W (B,B): w>1 for same-target negatives, 1 otherwise; diagonal left to CE."""
    B = len(grps)
    W = torch.ones((B,B), device=device)
    for i in range(B):
        for j in range(B):
            if i != j and grps[i] == grps[j]:
                W[i,j] = 2.0  # up-weight same-target impostors
    return W

def hardest_same_target_neg(logits, groups):
    """
    For each row i, find the best same-target negative j!=i.
    Return tensor of size [B] with the max logit among same-target negatives; if none, return -inf.
    """
    B = logits.size(0)
    hard = torch.full((B,), float("-inf"), device=logits.device)
    for i in range(B):
        mask = torch.tensor([ (groups[i]==groups[j] and j!=i) for j in range(B) ], device=logits.device)
        if mask.any():
            hard[i] = logits[i][mask].max()
    return hard

def contrastive_loss_with_weights(bt, bm, groups, T=0.07, margin=0.1):
    # Normalize
    bt = F.normalize(bt, dim=1)
    bm = F.normalize(bm, dim=1)
    logits = (bt @ bm.T) / T                      # [B,B]
    labels = torch.arange(bt.size(0), device=bt.device)

    # Weighted InfoNCE (row + col) with same-target emphasis
    W = build_same_target_weights(groups)         # [B,B]
    # Row-wise CE
    ce_row = F.cross_entropy(logits, labels, reduction='none')   # [B]
    # Column-wise CE
    ce_col = F.cross_entropy(logits.T, labels, reduction='none') # [B]
    # Weight by average same-target pressure (normalize to keep scales reasonable)
    # We multiply per-example CE by (1 + average weight on its negatives - 1) = average(W[i,neg])
    with torch.no_grad():
        B = logits.size(0)
        avg_w_row = (W.sum(dim=1) - torch.diag(W)) / (B-1)   # [B]
        avg_w_col = (W.sum(dim=0) - torch.diag(W)) / (B-1)   # [B]
    loss_row = (ce_row * avg_w_row).mean()
    loss_col = (ce_col * avg_w_col).mean()
    loss_nce = 0.5*(loss_row + loss_col)

    # Margin loss vs. hardest same-target impostor
    pos = logits.diag()                                      # [B]
    hard_same = hardest_same_target_neg(logits, groups)      # [B] (−inf if none)
    # hinge only where a same-target negative exists
    mask = torch.isfinite(hard_same)
    margin_loss = torch.clamp(margin - (pos[mask] - hard_same[mask]), min=0).mean() if mask.any() else torch.tensor(0.0, device=logits.device)

    return loss_nce + margin_loss, logits
```

```python

# Train
set_seed(SEED)
for ep in range(1, EPOCHS+1):
    proj_text.train(); proj_mol.train()
    running, n = 0.0, 0
    for Zt_b, Xm_b, grp_b, idx_b in tqdm(train_dl, desc=f"Epoch {ep}/{EPOCHS}"):
        Zt_b = Zt_b.to(device); Xm_b = Xm_b.to(device)
        Bt = proj_text(Zt_b);   Bm = proj_mol(Xm_b)

        loss, _ = contrastive_loss_with_weights(Bt, Bm, grp_b, T=TEMP, margin=0.1)
        opt.zero_grad(); loss.backward(); opt.step()

        running += loss.item(); n += 1
    print(f"Epoch {ep}: loss={running/max(n,1):.4f}")

# =============================
# 5) Evaluate on TEST split
# =============================
@torch.no_grad()
def embed_side(Z_text, X_mol):
    Bt = F.normalize(proj_text(torch.tensor(Z_text, dtype=torch.float32).to(device)), dim=1).cpu()
    Bm = F.normalize(proj_mol (torch.tensor(X_mol,  dtype=torch.float32).to(device)), dim=1).cpu()
    return Bt, Bm

Bt_test, Bm_test = embed_side(Z_text_test.numpy(), X_mol_test)
S = (Bt_test @ Bm_test.T)  # [N,N]
S_np = S.numpy()

# Global metrics
gt = np.arange(S_np.shape[0])
top1 = S_np.argmax(axis=1)
recall1 = (top1 == gt).mean()

ranks = []
for i in range(S_np.shape[0]):
    order = np.argsort(-S_np[i])
    rank_pos = int(np.where(order == i)[0][0]) + 1
    ranks.append(1.0 / rank_pos)
mrr = float(np.mean(ranks))

print(f"\n=== TEST (Scaffold split) ===")
print(f"Global Recall@1: {recall1:.3f}")
print(f"Global MRR:      {mrr:.3f}")

# Grouped (within-target) Recall@1
gR1 = grouped_recall_at1(S_np, test_df, group_col="target_chembl_id", min_group=MIN_GROUP_SIZE)
print(f"Grouped Recall@1 (within-target): {gR1:.3f}")
```

```python

import os, math, random, numpy as np, pandas as pd, torch, torch.nn as nn, torch.nn.functional as F
from tqdm import tqdm
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Scaffolds import MurckoScaffold
from transformers import AutoTokenizer, AutoModel
from torch.utils.data import Dataset, DataLoader

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

CSV_PATH = "chembl_mechanisms_enriched.csv"
TEXT_MODEL = "pritamdeka/S-Biomed-Roberta-snli-multinli-stsb"
ECFP_BITS, ECFP_RADIUS = 2048, 2
MAX_TXT_LEN = 96
SHARED_D = 256
BATCH_SIZE = 512
EPOCHS = 4          # quick; use 6–10 for final
WEIGHT_DECAY = 1e-4
LR = 1e-3
SEED = 0
MIN_GROUP_SIZE = 3

def set_seed(s=0):
    random.seed(s); np.random.seed(s); torch.manual_seed(s)
    if torch.cuda.is_available(): torch.cuda.manual_seed_all(s)

def ecfp4(smi, bits=ECFP_BITS, radius=ECFP_RADIUS):
    m = Chem.MolFromSmiles(smi)
    if m is None: return np.zeros(bits, dtype=np.float32)
    bv = AllChem.GetMorganFingerprintAsBitVect(m, radius, nBits=bits)
    arr = np.zeros((bits,), dtype=np.int8)
    Chem.DataStructs.ConvertToNumpyArray(bv, arr)
    return arr.astype(np.float32)

def murcko(sm):
    m = Chem.MolFromSmiles(sm)
    if m is None: return None
    core = MurckoScaffold.GetScaffoldForMol(m)
    return Chem.MolToSmiles(core) if core else None

@torch.no_grad()
def cls_encode(texts, tok, model, max_length=96, bs=64):
    outs = []
    for i in range(0, len(texts), bs):
        enc = tok(texts[i:i+bs], padding=True, truncation=True, max_length=max_length, return_tensors="pt").to(device)
        outs.append(model(**enc).last_hidden_state[:,0,:].detach().cpu())
    return torch.cat(outs, dim=0)

def grouped_recall_at1(S_np, meta_df, group_col="target_chembl_id", min_group=3):
    gt = np.arange(S_np.shape[0])
    # compute within-group top-1
    groups = meta_df[group_col].fillna("UNK").astype(str).values
    idx = {}
    for i,g in enumerate(groups):
        idx.setdefault(g, []).append(i)
    hits = total = 0
    for g, idxs in idx.items():
        if len(idxs) < min_group: continue
        sub = S_np[np.ix_(idxs, idxs)]
        top1 = sub.argmax(axis=1)
        hits += (top1 == np.arange(len(idxs))).sum()
        total += len(idxs)
    return (hits/total) if total else 0.0

def load_split(remove_drug_name=False):
    df = pd.read_csv(CSV_PATH).dropna(subset=["smiles","text_rich"]).copy()
    if remove_drug_name:
        df["text_rich"] = df["text_rich"].str.replace(r"Drug:\s*[^.]+\. ?", "", regex=True).str.strip()
    df["scaffold"] = df["smiles"].apply(murcko)
    df = df.dropna(subset=["scaffold"]).reset_index(drop=True)
    scafs = df["scaffold"].unique()
    np.random.seed(SEED); np.random.shuffle(scafs)
    cut = int(0.8*len(scafs))
    train_df = df[df["scaffold"].isin(set(scafs[:cut]))].reset_index(drop=True)
    test_df  = df[df["scaffold"].isin(set(scafs[cut:]))].reset_index(drop=True)

    # features
    X_train = np.stack([ecfp4(s) for s in train_df["smiles"].tolist()])
    X_test  = np.stack([ecfp4(s) for s in test_df["smiles"].tolist()])

    tok = AutoTokenizer.from_pretrained(TEXT_MODEL, use_fast=True)
    model = AutoModel.from_pretrained(TEXT_MODEL).to(device).eval()
    Zt_train = cls_encode(train_df["text_rich"].tolist(), tok, model, max_length=MAX_TXT_LEN, bs=64)
    Zt_test  = cls_encode(test_df["text_rich"].tolist(),  tok, model, max_length=MAX_TXT_LEN, bs=64)

    return train_df, test_df, X_train, X_test, Zt_train, Zt_test

class PairDS(Dataset):
    def __init__(self, Zt, Xm, groups):
        self.Zt = torch.tensor(Zt, dtype=torch.float32)
        self.Xm = torch.tensor(Xm, dtype=torch.float32)
        self.grp = np.array(groups).astype(str)
    def __len__(self): return len(self.Zt)
    def __getitem__(self, i): return self.Zt[i], self.Xm[i], self.grp[i]

def train_and_eval(train_df, test_df, Zt_tr, Xm_tr, Zt_te, Xm_te, T=0.07, margin=0.1):
    ds = PairDS(Zt_tr.numpy(), Xm_tr, train_df["target_chembl_id"].tolist())
    dl = DataLoader(ds, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
    proj_t = nn.Linear(Zt_tr.shape[1], SHARED_D).to(device)
    proj_m = nn.Linear(Xm_tr.shape[1], SHARED_D).to(device)
    opt = torch.optim.AdamW(list(proj_t.parameters())+list(proj_m.parameters()), lr=LR, weight_decay=WEIGHT_DECAY)

    def build_same_target_weights(grps):
        B = len(grps); W = torch.ones((B,B), device=device)
        for i in range(B):
            for j in range(B):
                if i!=j and grps[i]==grps[j]: W[i,j]=2.0
        return W

    def hardest_same_target_neg(logits, groups):
        B = logits.size(0)
        hard = torch.full((B,), float("-inf"), device=logits.device)
        for i in range(B):
            mask = torch.tensor([ (groups[i]==groups[j] and j!=i) for j in range(B) ], device=logits.device)
            if mask.any(): hard[i] = logits[i][mask].max()
        return hard

    for ep in range(1, EPOCHS+1):
        proj_t.train(); proj_m.train()
        tot = n = 0
        for Zt_b, Xm_b, grp_b in dl:
            Zt_b = Zt_b.to(device); Xm_b = Xm_b.to(device)
            Bt = F.normalize(proj_t(Zt_b), dim=1)
            Bm = F.normalize(proj_m(Xm_b), dim=1)
            logits = (Bt @ Bm.T) / T
            labels = torch.arange(Bt.size(0), device=Bt.device)

            W = build_same_target_weights(grp_b)
            ce_row = F.cross_entropy(logits, labels, reduction='none')
            ce_col = F.cross_entropy(logits.T, labels, reduction='none')
            with torch.no_grad():
                B = logits.size(0)
                avg_w_row = (W.sum(dim=1) - torch.diag(W)) / (B-1)
                avg_w_col = (W.sum(dim=0) - torch.diag(W)) / (B-1)
            loss_nce = 0.5*((ce_row*avg_w_row).mean() + (ce_col*avg_w_col).mean())

            pos = logits.diag()
            hard_same = hardest_same_target_neg(logits, grp_b)
            mask = torch.isfinite(hard_same)
            loss_margin = torch.clamp(margin - (pos[mask] - hard_same[mask]), min=0).mean() if mask.any() else 0.0

            loss = loss_nce + loss_margin
            opt.zero_grad(); loss.backward(); opt.step()
            tot += float(loss.item()); n += 1

    # eval
    @torch.no_grad()
    def embed(Zt, Xm):
        Bt = F.normalize(proj_t(torch.tensor(Zt, dtype=torch.float32).to(device)), dim=1).cpu()
        Bm = F.normalize(proj_m(torch.tensor(Xm, dtype=torch.float32).to(device)), dim=1).cpu()
        S = (Bt @ Bm.T).numpy()
        gt = np.arange(S.shape[0])
        top1 = S.argmax(axis=1)
        recall1 = (top1==gt).mean()
        ranks = []
        for i in range(S.shape[0]):
            order = np.argsort(-S[i]); r = int(np.where(order==i)[0][0]) + 1
            ranks.append(1.0/r)
        mrr = float(np.mean(ranks))
        gR1 = grouped_recall_at1(S, test_df, "target_chembl_id", min_group=MIN_GROUP_SIZE)
        return recall1, mrr, gR1

    return embed(Zt_te.numpy(), Xm_te)

# Run sweep
grid_T = [0.05, 0.07, 0.10]
grid_m = [0.05, 0.10, 0.15]
grid_leak = [False, True]  # False = with drug name; True = strip drug name

results = []
for rm_drug in grid_leak:
    print("\nPreparing data (remove_drug_name =", rm_drug, ") ...")
    set_seed(SEED)
    train_df, test_df, X_tr, X_te, Zt_tr, Zt_te = load_split(remove_drug_name=rm_drug)
    for T in grid_T:
        for m in grid_m:
            set_seed(SEED)
            r1, mrr, gR1 = train_and_eval(train_df, test_df, Zt_tr, X_tr, Zt_te, X_te, T=T, margin=m)
            results.append({"remove_drug_name": rm_drug, "T": T, "margin": m,
                            "Global_R1": r1, "MRR": mrr, "Grouped_R1": gR1})
            print(f"rm_drug={rm_drug} T={T} m={m}  ->  R@1={r1:.3f}  MRR={mrr:.3f}  gR@1={gR1:.3f}")

res = pd.DataFrame(results).sort_values(["remove_drug_name","Grouped_R1","Global_R1"], ascending=[True,False,False])
print("\nAblation summary (top rows):")
display(res.head(10))
```
